```python
import wooldridge as woo
import statsmodels.formula.api as smf
import numpy as np
```

```python
gpa1 = woo.dataWoo('gpa1')
gpa1.head()
```

```python
reg_gap1 = smf.ols(formula='colGPA ~ hsGPA + ACT', data=gpa1)
results_gap1 = reg_gap1.fit()
print(f"results_gap1.summary():\n{results_gap1.summary()}\n")
```

wage数据集

```python
wage1 = woo.dataWoo('wage1')

reg_wage = smf.ols(formula='np.log(wage) ~ educ + exper + tenure', data=wage1)
results_wage = reg_wage.fit()
print(f"results_wage.summary():\n{results_wage.summary()}\n")
```

401K数据集

```python
k401k = woo.dataWoo('401k')

reg_401k = smf.ols(formula='prate ~ mrate + age', data=k401k)
results_401k = reg_401k.fit()
print(f"results_401k.summary():\n{results_401k.summary()}\n")
```

crime数据集

```python
crime1 = woo.dataWoo('crime1')

reg_crime = smf.ols(formula='narr86 ~ pcnv + ptime86 + qemp86', data = crime1)
results_crime = reg_crime.fit()
print(f"results_crime.summary():\n{results_crime.summary()}\n")
```
